```python
import pandas as pd
```

```python
base_path = "../00_datasets/dataset_output/include50/"
```

```python
with open(base_path + "include50_train.txt") as file:
    train_signs = [line.rstrip() for line in file]
```

```python
with open(base_path + "include50_test.txt") as file:
    test_signs = [line.rstrip() for line in file]
```

```python
with open(base_path + "include50_val.txt") as file:
    val_signs = [line.rstrip() for line in file]
```

```python
df = pd.read_csv(base_path + "include50_openpose.csv")
```

```python
df = df[df["frame"] == 0]
```

```python
df["video_path"] = df.apply(lambda i: (i["video_name"].replace("\\", "/")).lower().replace(".mov", "").replace(".mp4", ""), axis=1)
```

```python
videos = df["video_path"].unique()
videos
```

```python
for video in train_signs:
    if video.lower().replace(".mov", "").replace(".mp4", "") not in videos:
        print(video)
```

```python
for video in test_signs:
    if video.lower().replace(".mov", "").replace(".mp4", "") not in videos:
        print(video)
```

```python
for video in val_signs:
    if video.lower().replace(".mov", "").replace(".mp4", "") not in videos:
        print(video)
```
